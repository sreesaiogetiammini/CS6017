# Part 1 Python/Numpy Warmup

```python
import numpy as np
rand_array = np.random.rand(100)
def compute_mean(data):
    return np.sum(data)/len(data)
def compute_standart_deviation(data):
    mean = compute_mean(data)
    squared_difference = np.square(data - mean)
    variance = np.sum(squared_difference)/(len(data) - 1)
    std_deviation = np.sqrt(variance)
    return std_deviation
mean = compute_mean(rand_array)
standard_deviation = compute_standart_deviation(rand_array)

print("Mean for 100 random numbers: ", mean)
print("Standard Deviation for 100 random numbers: ", standard_deviation)
```

```python
rand_array = np.random.rand(100000)
mean = compute_mean(rand_array)
standard_deviation = compute_standart_deviation(rand_array)

print("Mean for 100000 random numbers: ", mean)
print("Standard Deviation for 100000 random numbers: ", standard_deviation)
```

The mean is almost the same for both the samples. The standard deviation which almost closeto the standard deviation of a uniform distribution between 0 and 1. 

The larger the sample size provides a more accurate estimate of the population parameters.

```python
import scipy as spy
import matplotlib.pyplot as plt

rand_array = spy.stats.norm.rvs(size=10000)
mean = compute_mean(rand_array)
standard_dev = compute_standart_deviation(rand_array)
mean_builtIn = np.mean(rand_array)
standard_deviation_builtIn = np.std(rand_array)

print("Mean for 10000 random numbers: ", mean)
print("Standard Deviation for 10000 random numbers: ", standard_deviation)
print("Built in Mean for 10000 random numbers: ", mean_builtIn)
print("Built in Standard Deviation for 10000 random numbers: ", standard_deviation_builtIn)
```

```python
plt.hist(rand_array, bins=10, alpha=0.5, label='10 bins', color="red")
plt.hist(rand_array, bins=20, alpha=0.5, label='20 bins', color="blue")
plt.hist(rand_array, bins=40, alpha=0.5, label='40 bins', color="yellow")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()
plt.show()
```

The histogram of the samples is a bell-shaped curve, which is the characteristic shape of a normal distribution. Increasing the number of bins in the histogram provides more granularity in the visualization, allowing for better analysis of the distribution.

# Part 2 Data Exploration/Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./2021_PM_2.5.csv')
station_name = 'CV-MC'
plt.plot(data['Date'], data['CV-MC'])
plt.xlabel('Date')
plt.ylabel('PM2.5')
plt.title(f'Hourly PM2.5 Readings - {station_name}')
plt.xticks(rotation=45)
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
monthly_mean = data.groupby(data['Date'].dt.month)['CV-MC'].mean()
plt.bar(monthly_mean.index, monthly_mean)
plt.xlabel('Month')
plt.ylabel('Mean PM2.5')
plt.title(f'Mean PM2.5 Level by Month - {station_name}')
plt.xticks(range(1, 13))
plt.show()

# plt.boxplot(box_and_whisker_monthly.values)
# plt.xlabel('Month')
# plt.ylabel('Value')
# plt.title(f'Monthly Data Box Plot at- {station_name}')
# plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
# plt.show()
```

The highest average pollution is recorded during August in 2021 and the least pollution is recorded in February.

```python
hourly_mean = data.groupby(data['Date'].dt.hour)['CV-MC'].mean()
plt.plot(hourly_mean.index, hourly_mean)
plt.xlabel('Hour')
plt.ylabel('Mean PM2.5')
plt.title(f'Mean PM2.5 Level by Hour - {station_name}')
plt.xticks(range(24))
plt.show()
```

This visualization gives the hourly average pollution recorded during the day throughout the year and it is least during the afternoons aroung 3pm to 4pm and the highest is during the nights.

```python
monthly_boxplot = data.boxplot(column='CV-MC', by=data['Date'].dt.month)
plt.xlabel('Month')
plt.ylabel('PM2.5')
plt.title(f'Box and Whisker Plot of PM2.5 Levels by Month - {station_name}')
plt.xticks(range(1, 13))
plt.suptitle('')
plt.show()
```

```python
hourly_boxplot = data.boxplot(column='CV-MC', by=data['Date'].dt.hour)
plt.xlabel('Hour')
plt.ylabel('PM2.5')
plt.title(f'Box and Whisker Plot of PM2.5 Levels by Hour - {station_name}')
plt.xticks(range(24))
plt.suptitle('')
plt.show()
```

This box and whisker plot here gives us the insights about the extremes for a average day. It also gives where the first 25% of the data falls. This also gives us data about the outliers.
